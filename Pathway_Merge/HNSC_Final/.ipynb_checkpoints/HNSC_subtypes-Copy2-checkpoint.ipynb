## Setup Survival Screen Functions

```python
from Processing.Screen import *
```

```python
class Screen(object):
    '''
    Object to hold data and results for survival screen.
    '''
    
    def __init__(self, mut, cn, rna, mirna, clinical_df):       
        
        '''Process Gene / Pathway Expression'''
        pathways = remove_redundant_pathways(rna.pathways, rna.global_vars.background, 
                                             binarize=True)
        rna_gene_df = extract_diff_exp_rna(rna, n=300, binarize=True)
        self.rna_df = pd.concat([rna_gene_df, pathways]).T
        
        '''Process miRNA Expression'''
        mirna_binarized = mirna.features.ix['real'].apply(binarize_feature, 1)
        self.mirna_df = pd.concat([mirna.features.ix['binary'], mirna_binarized]).T
        
        '''Process mutation data'''
        self.rate = mut.df.sum()
        self.mut_df = mut.features
        
        '''Process CNA data'''
        self.cna_df = cn.features
        
        '''Process Clinical Data'''
        self.clinical_df  = clinical_df
        
    def get_patient_set(self, filters):
        f1 = list(filters)
        filter_df = pd.concat(f1, axis=1)
        clinical_filter = filter_df.dropna().sum(1) == 0
        keepers_o = true_index(clinical_filter)
        keepers_o = keepers_o.intersection(self.mut_df.columns)
        keepers_o = keepers_o.intersection(self.cna_df.columns)
        return keepers_o
        
    def get_data(self, keepers_o, cutoff=12):
        mut_df = mut_filter(self.mut_df.ix[:,keepers_o], self.rate, 
                            cutoff).T
        amp_df, del_df = cn_filter(self.cna_df.ix[:, keepers_o], cutoff)
        cna_df = pd.concat([del_df, amp_df], keys=['del','amp'], axis=1)
        cna_df.columns = map(lambda i: '_'.join(i), cna_df.columns)
        
        df = pd.concat({'clinical': self.clinical_df,
                        'mutation': mut_df,
                        'cna': cna_df,
                        'rna': self.rna_df,
                        'mirna': self.mirna_df}, axis=1).T
        df = df.ix[:, keepers_o]
        df = filter_binary(df, cutoff)
        return df
```

```python
def test(s, surv, cov_df):
    s = s.dropna()
    try:
        return get_cox_ph_ms(surv, s, cov_df, return_val='LR', interactions='just_feature')
    except:
        return pd.Series(index=['LR','feature_p', 'fmla', 'hazzard'])


def run_screen(screen, filters, covariates):
    cov_df = pd.concat(covariates, axis=1)
    keepers_o = screen.get_patient_set(filters)
    cutoff = max(np.ceil(len(keepers_o) * .05), 10)
    df = screen.get_data(keepers_o, cutoff)

    univariate = cox_screen(df, surv)
    vec = univariate.LR.p.sort_index()
    univariate['q_bonf'] = vec.groupby(level=0).apply(lambda s: s*len(s)).order()
    univariate['q_bh'] = vec.groupby(level=0).apply(bhCorrection).order()
    #hits = univariate[univariate['q_bh'] < .2].index
    hits = univariate.index
    
    feature_age = df.ix[hits].apply(test, args=(surv, cov_df,), axis=1)
    vec = feature_age.LR.ix[univariate.index].sort_index()
    feature_age['q_bonf_within'] = vec.groupby(level=0).apply(lambda s: s*len(s)).order()
    feature_age['q_bh_within'] = vec.groupby(level=0).apply(bhCorrection).order()
    feature_age['q_two_step'] = feature_age['q_bh_within'] * 4
    feature_age['q_bh_all'] = bhCorrection(vec)
    feature_age['q_bonf_all'] = vec * len(vec)
    hits = true_index(feature_age.sort('LR').q_bh_all < .1)

    try:
        pairs, interactions = get_interactions(df.ix[hits], cov_df, 
                                               surv, test)
        res = pd.concat([feature_age, interactions]).sort('LR')
    except:
        res = feature_age.sort('LR')
        pairs = []
    return res, pairs, feature_age, univariate, keepers_o, df
```

```python
s = Screen(mut, cn, rna, mirna, clinical_df)
```

```python
res, pairs, feature_age, univariate, keepers_o, df = run_screen(s, [hpv_inferred==1], [old, age])
```
